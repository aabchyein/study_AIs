```python
import pymongo as mg
import pandas as pd
```

```python
client = mg.MongoClient(host='mongodb://localhost:27017')
database = client['db_NHIS']
collection = database['Join Collection NSC2_BND_M20']
cursor = collection.find({})
list_BNS_M20 = list(cursor)
df_BNS_M20 = pd.DataFrame(list_BNS_M20)
df_BNS_M20[:2]
```

#### 전처리
- SICK_SYM1은 상병을 나타냄 -> 3자리 대분류로 자르기
- F_, N_ 값들은 NULL 값으로 만들기 : 상병에 맞지 않는 값들이므로 자료를 없애주는 것

```python
df_BNS_M20['SICK_SYM1'].value_counts()
```

```python
len('J209'), len('F_')
```

```python
'J209'[:3]
# str('J209')[:3]
```

```python
# sick_sym에 담기는 값들: J209, I10, I109...
def convertSICK_SYM(sick_sym) :
    if len(sick_sym) < 3 :
        return pd.NA
    else :
        return sick_sym[:3]
# return을 중간에 조건이 맞으면 쓰기도 함!
```

```python
convertSICK_SYM('J209'), convertSICK_SYM('F_')
```

```python
# apply()적용해서 새로운 컬럼 만들기
df_BNS_M20['SICK_SYM1_3'] = df_BNS_M20['SICK_SYM1'].apply(convertSICK_SYM)
```

```python
df_BNS_M20['SICK_SYM1_3'].value_counts()
```

```python
# null값 삭제
df_BNS_M20_dropna = df_BNS_M20.dropna()
df_BNS_M20_dropna.isnull().sum()
```

```python
df_BNS_M20_dropna[:2]
```

```python
target_list = df_BNS_M20_dropna['SICK_SYM1_3'].value_counts().index.to_list()
```

```python
target_list.index('L02'), target_list.index('J03')
```

```python
# SICK_SYM1_3 값을 넣었을 때 index값, 즉 어느 위치에 있는지 알고자 함
def applyIndexNumber(sick_sym_3):
    indexNumber = target_list.index(sick_sym_3)
    return indexNumber
```

```python
df_BNS_M20_dropna['SICK_SYM1_3_target'] = df_BNS_M20_dropna['SICK_SYM1_3'].apply(applyIndexNumber)
df_BNS_M20_dropna[:2]
```

### 정형화

```python
df_BNS_M20_dropna.columns
```

```python
target = df_BNS_M20_dropna['SICK_SYM1_3_target']
features = df_BNS_M20_dropna[['EDC_ADD_RT', 'ED_RC_TOT_AMT']]
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=111)
features_train.shape, target_train.shape, features_test.shape, target_test.shape
```

#### 모델학습
- 목표변수(범주형): DecisionTreeClassifier
- 목표변수(연속형): DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
model = DecisionTreeClassifier()
```

```python
# sklearn의 GridSearchCV는 머신러닝 알고리즘에 사용되는
# 하이퍼 파라미터를 입력해 학습을 하고 측정을 하면서 가장 좋은 파라미터를 알려줌
from sklearn.model_selection import GridSearchCV
```

```python
# param_griddict or list of dictionaries
# 튜닝을 위해 사용될 파라미터를 dictionary 형태로 만들어서 넣는다
hyper_params = {'min_samples_leaf' : range(2,5)
                , 'max_depth' : range(2,5)
                , 'min_samples_split' : range(2,5)}

# range(2,5) -> [2,3,4,5]
```

##### 평가
- score : 예측 성능을 측정할 평가 방법을 넣을 수 있다.
    * defalut값: 분류-accuracy, 예측-R square
    * 분류-정확도(accuracy), f1
    * 예측-R square

```python
from sklearn.metrics import f1_score, make_scorer
```

```python
# 변경하고자 하는 평가 방식 넣기
# best_estimator_를 할 때 default값인 accuracy가 아닌 f1값으로 찾아줌
scoring = make_scorer(f1_score)
```

```python
# 실험하려는 모델에 하이퍼파라미터와 값 범위를 지정하면
# GridSearchCV는 교차검증을 사용하여 하이퍼파라미터 값의 가능한 모든 조합을 수행
grid_search = GridSearchCV(model, param_grid=hyper_params, cv=3
                           , verbose=1, scoring=scoring)

# grid_search에는 3x2x3 에 cv의 값을 곱한 54개가 들어있다.
```

```python
grid_search.fit(features, target)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_, grid_search.best_params_

# best_score_ : 0.14814814814814814 (accuracy;정확도가 떨어짐)
# best_score_ : nan (값이 너무 작기 때문)
```

```python
# 54개 중 최고의 성능을 가진 하나를 찾아서 변수에 담아줌
best_model = grid_search.best_estimator_
best_model
```

```python
target_test_predict = best_model.predict(features_test)
target_test_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target_test, target_test_predict))
```
